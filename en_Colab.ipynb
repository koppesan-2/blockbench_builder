## Selective Build

```python
# @title Specify the build destination and execute the build.
ostype = "--linux" # @param ["--linux",""] {"allow-input":true}
architecture = "--arm64" # @param ["--x64","--arm64","--armv7l","--ia32",""] {"allow-input":true}
filetype = "deb" # @param ["AppImage","deb","rpm","freebsd","pacman","apk","--universal","--dir",""] {"allow-input":true}
# @markdown **Description of Each Option**
# @markdown * ostype→Specifies the target OS type for installation
# @markdown * architecture→Specifies the target PC's architecture (usually selecting x64 is fine)
# @markdown * filetype→Specifies the file type of the generated output
# @markdown
# @markdown Leaving any option blank will build using the default settings (for Linux x64)
# @markdown
# @markdown While impossible combinations are possible, they will cause errors and halt the process
# @markdown
# @markdown Builds take approximately 1 to 15 minutes without a paid subscription, depending on your choices. Please be patient.
%cd /content/blockbench/
!npx electron-builder $ostype $filetype $architecture
```
