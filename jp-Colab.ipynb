# 実際にビルドする

## 一部ビルド向けの追加の前提パッケージインストール

```python
# @title pacmanビルド時用追加パッケージインストール
%cd /content/blockbench
!sudo apt install libarchive-tools
```

## 選択型ビルド

```python
# @title ビルド先を指定してビルド
ostype = "--linux" # @param ["--linux",""] {"allow-input":true}
architecture = "--arm64" # @param ["--x64","--arm64","--armv7l","--ia32",""] {"allow-input":true}
filetype = "deb" # @param ["AppImage","deb","rpm","freebsd","pacman","apk","--universal","--dir",""] {"allow-input":true}
# @markdown **それぞれの選択肢の説明**
# @markdown * ostype→インストール先のOSタイプを指定
# @markdown * architecture→インストール先のPCのアーキテクチャを指定(大体はx64を選んでおけばok)
# @markdown * filetype→成果物のファイルタイプ
# @markdown
# @markdown それぞれの選択肢で空欄を選ぶとデフォルト(linux x64向け)の設定でビルドが実行されます
# @markdown
# @markdown ありえない選択の組み合わせもできますが、エラーが出て、止まります
# @markdown
# @markdown ビルドには、無課金の場合、選択肢にもよりますが、1~15分程度はかかるので気長にお待ちください
%cd /content/blockbench/
!npx electron-builder $ostype $filetype $architecture
```

# プリセットとして用意しておいたビルド方法たち
非表示セルを表示すると、プリセットとして実行ボタンを押すだけで、ビルドができるようにしておきました

それぞれ、linux arm64向けと、linux x64向けのビルドをワンボタンで生成します

(無課金の場合、8~10分程度かかります)

```python
# @title Linux arm64向けビルド (成果品はRaspberryPi5で動作することを確認済み)
!npx electron-builder --linux --arm64
# @title Linux x64向けビルド (x64系のコンピュータを持っていないので、動作確認はできていません)

!npx electron-builder --linux --x64
```

```python
# @title Linux x64向けビルド (x64系のコンピュータを持っていないので、動作確認はできていません)

!npx electron-builder --linux --x64
```

それぞれのファイル名
*   Appimage → Blockbench_{version}.AppImage
*   debファイル → Blockbench_{version}.deb
*   rpmファイル → Blockbench_{version}.rpm

ファイルを右クリック→ダウンロード　で成果物をダウンロードしてください

あとは、

(debian,ubuntuの場合)
> sudo apt install <成果物へのパス>
>/opt/Block

(AppImageの場合)
>./<成果物へのパス>
※AppImageは、
>chmod +x <成果物へのパス>
などで、実行権限の付与が必要かもしれません
